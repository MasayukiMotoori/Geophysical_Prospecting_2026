```python
# --- helper to render LaTeX labels (robust method) ---
def math_label(tex, width="160px"):
    out = widgets.Output(layout=widgets.Layout(width=width))
    with out:
        display(Math(tex))
    return out
def sci_latex(v, prec=2):
    s = f"{v:.{prec}e}"          # e.g. '3.00e-03'
    mant, exp = s.split('e')
    exp = int(exp)

    if float(mant) == 0:
        return "0"

    if exp in [0, -1]:
        return f"{float(mant)*10**exp:.{prec}f}"
    else:
        return rf"{mant}\cdot 10^{{{exp}}}"


def fmt(v, prec=2, latex=False, wrap=False):
    # blank for None or empty string
    if v is None or v == "":
        return ""

    # if already string → return as-is
    if isinstance(v, str):
        return v

    try:
        s = sci_latex(v, prec=prec)

        if latex:
            if wrap:
                return rf"$${s}$$"   # display math
            else:
                return rf"${s}$"     # inline math
        else:
            return s

    except:
        return ""
```

```python
def plot_vessel_outline(ax, linewidth=0.4, color='k', zorder=5):
    # Outer vessel
    x_out = width_vessel * np.r_[-1, 1, 1, -1, -1]
    y_out = height_vessel * np.r_[1, 1, 0, 0, 1] + frame_z

    # Inner vessel
    x_in = (width_vessel - thick_vessel) * np.r_[-1, 1, 1, -1, -1]
    y_in = ((height_vessel - 2 * thick_vessel) * np.r_[1, 1, 0, 0, 1]
            + thick_vessel + frame_z)

    ax.plot(x_out, y_out, color=color, linewidth=linewidth,
            zorder=zorder, solid_joinstyle='miter')
    ax.plot(x_in, y_in, color=color, linewidth=linewidth,
            zorder=zorder, solid_joinstyle='miter')

    return ax
def plot_frame_outline(ax, linewidth=0.4, color='k', zorder=5):
    x = width_frame * np.r_[-1, 1, 1, -1, -1]
    y = height_frame * np.r_[1, 1, 0, 0, 1]

    ax.plot(x, y, color=color, linewidth=linewidth,
            zorder=zorder, solid_joinstyle='miter')

    return ax
```

## Set Up mapping

```python


ind_container = (
    (mesh.cell_centers[:, 2] <= height_vessel + frame_z) &
    (mesh.cell_centers[:, 2] >= frame_z) &
    (mesh.cell_centers[:, 0] <= width_vessel)
)

ind_sea = (mesh.cell_centers[:, 2] > - tx_height) & ~ind_container
ind_active = ~ind_sea
ind_vessel = (
    (mesh.cell_centers[ind_active][:, 2] <= height_vessel + frame_z) &
    (mesh.cell_centers[ind_active][:, 2] >= frame_z) &
    (mesh.cell_centers[ind_active][:, 0] <= width_vessel)
)
ind_air = (
    (mesh.cell_centers[ind_active][:, 2] < height_vessel - thick_vessel + frame_z) &
    (mesh.cell_centers[ind_active][:, 2] > thick_vessel + frame_z) &
    (mesh.cell_centers[ind_active][:, 0] < width_vessel - thick_vessel)
)

ind_metal = ind_vessel &  ~ind_air
ind_layer = ((mesh.cell_centers[ind_active][:, 2] <= - tx_height) & 
            (mesh.cell_centers[ind_active][:, 2] > -( tx_height + thick_target)) )
ind_seabed = mesh.cell_centers[ind_active][:, 2] <= -( tx_height + thick_target)
```

```python
print(mesh.n_cells)
print(ind_active.sum())
print(ind_vessel.sum())
print(ind_air.sum())
print(ind_metal.sum())
print(ind_layer.sum())
print(ind_seabed.sum())
assert ind_active.sum() == ind_air.sum() + ind_metal.sum() + ind_layer.sum() + ind_seabed.sum()
```

```python
indices_list = [ind_metal, ind_air, ind_layer, ind_seabed]
active_map = maps.InjectActiveCells(mesh= mesh,
             active_cells=ind_active  , value_inactive=np.log(sigma_sea))
projection_map = maps.SurjectUnits(indices_list)
## Kill indices for the seawater

## Exponential map to convert the model to conductivity
exp_map = maps.ExpMap(mesh)
model_map = exp_map*active_map *  projection_map
#model_map = exp_map*active_map *  parametric_map
```

```python
time_steps = [(2e-6,  150),  (5e-6,  150),  (2e-5,  150), (5e-5,  10)]
time_step_min = 2e-6
tstrt, tend = 1e-4, 4.0e-3
steps = np.array([])
for i in range(len(time_steps)):
    dt, n = time_steps[i]
    step_array = dt * np.ones(n)
    steps = np.concatenate((steps, step_array))
time_clc = np.cumsum(steps)
# since using step off waveform
rmp_time_sim = time_clc[0]
time_rec = time_clc[1:] -rmp_time_sim 
print(time_clc)
print(time_rec)
```

```python
def plot_model(model, clim=[-8, 6], xlim=[-3, 3], ylim=[-3., 3],
               ax=None, add_colorbar=False, cax=None,
               cbar_kwargs=None, cbar_label=r'$\log_{10}\sigma$'):
    if ax is None:
        fig, ax = plt.subplots(1, 1)

    # plot and return the mappable (img)
    img = mesh.plot_image(np.log10(model_map * model), clim=clim, ax=ax, mirror=True)[0]
    img.set_rasterized(True)  # rasterize the image for better performance
    # optional colorbar
    cb = None
    if add_colorbar:
        if cbar_kwargs is None:
            cbar_kwargs = {}
        # IMPORTANT: use fig.colorbar, not plt.colorbar
        fig = ax.figure
        cb = fig.colorbar(img, ax=ax if cax is None else None, cax=cax, **cbar_kwargs)
        cb.set_label(cbar_label)

    # extras
    ax.plot(np.r_[src.radius, -src.radius],
            np.r_[src.location[2], src.location[2]], "ko-", markersize=1.0)

    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_aspect('equal')
    ax.set_xlabel('x (m)')
    ax.set_ylabel('z (m)')
    ax.grid(which='both', color='k', linestyle='-', linewidth=0.5)

    return img, cb
```

```python

def apply_lowpass_fiilter_time(data, f_in, cutoff,order=1):
    nyquist = 0.5 * f_in
    normal_cutoff = cutoff / nyquist
    b,a = signal.butter(order, normal_cutoff, btype='low', analog=False)
    return signal.filtfilt(b,a,data)

def downsample(data, down):
    n_in = data.size
    n_out = n_in // down
    downsampled = np.zeros(n_out)
    for i in range(n_out):
        downsampled[i] = np.mean(data[i*down:(i+1)*down]) 
    return downsampled[:n_out]
```
